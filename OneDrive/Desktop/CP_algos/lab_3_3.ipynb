```python
import numpy as np
import matplotlib.pyplot as plt

def duffing(x, r, alpha, beta):
    return r - alpha*x - beta*x**3

x = np.linspace(-3, 3, 400)

r_values = [-1, 0, 1]

plt.figure(figsize=(10,4))

# Case 1: alpha > 0, beta < 0
plt.subplot(1,2,1)
alpha = 1
beta = -1
for r in r_values:
    dx = duffing(x, r, alpha, beta)
    plt.plot(x, dx, label=f"r={r}")

plt.axhline(0, color='black')
plt.title(">0, <0 (Bistable case)")
plt.xlabel("x")
plt.ylabel("dx/dt")
plt.legend()

# Case 2: alpha > 0, beta > 0
plt.subplot(1,2,2)
alpha = 1
beta = 1
for r in r_values:
    dx = duffing(x, r, alpha, beta)
    plt.plot(x, dx, label=f"r={r}")

plt.axhline(0, color='black')
plt.title(">0, >0 (Single stable point)")
plt.xlabel("x")
plt.legend()

plt.tight_layout()
plt.show()
```
